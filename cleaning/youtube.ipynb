```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
yt_data = pd.read_csv('ytube_results_4_of_156.csv')
yt_data.head()
```

```python
# Drop the urls
yt_data.drop(['url'], axis=1, inplace=True)
```

```python
# Read a column in full
pd.set_option('display.max_colwidth', None)
yt_data.head()
```

```python
# Clean the views column and convert to integer
def convert_to_int(x):
    try:
        # Remove 'views' and any whitespace
        x = x.replace('views', '').strip()
        
        # Handle K (thousands)
        if 'K' in x:
            return int(float(x.replace('K', '')) * 1000)
        # Handle M (millions)
        elif 'M' in x:
            return int(float(x.replace('M', '')) * 1000000)
        # Handle plain numbers
        else:
            return int(x.replace(',', ''))
    except:
        # Return 0 for any invalid or 'No' views
        return 0

yt_data['views'] = yt_data['views'].apply(convert_to_int)

yt_data.head()
```

```python
# The title column values include the channel name, let's split them at the '|' character
# We already have the channel name in the channel column
# For some columns, the character is more than one, so we will split at the first occurrence
yt_data['title'] = yt_data['title'].apply(lambda x: x.split('|')[0])

yt_data.head()
```

```python
# Function to convert age to months
def convert_to_months(age_str):
    try:
        # Extract number and unit
        number = int(age_str.split()[0])
        if 'year' in age_str:
            return number * 12
        elif 'month' in age_str:
            return number
        else:
            return None
    except:
        return None

# Convert age to months
yt_data['age'] = yt_data['age'].apply(convert_to_months)

# Rename column
yt_data = yt_data.rename(columns={'age': 'age_months'})

# Verify conversion
print(yt_data['age_months'].head())
```

```python
yt_data.head()
```

```python
# the scrape_date column is a timestamp, let's convert it to a datetime object
yt_data['scrape_date'] = pd.to_datetime(yt_data['scrape_date'])

yt_data.head()
```

```python
# From the scrape_date column, we can extract the year, month, and day
yt_data['scrape_year'] = yt_data['scrape_date'].dt.year
yt_data['scrape_month'] = yt_data['scrape_date'].dt.month
yt_data['scrape_day'] = yt_data['scrape_date'].dt.day

yt_data.head()
```

```python
# Drop the scrape_date column
#yt_data.drop(['scrape_date'], axis=1, inplace=True)

# save the cleaned data
yt_data.to_csv('ytube_results_4_of_156_cleaned.csv', index=False)
```
