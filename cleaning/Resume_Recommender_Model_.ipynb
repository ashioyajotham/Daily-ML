Functions to get skills from the entity ruler from the JSONL file, loading the spaCy English Model, removing duplicate skills and loading the resumes to be used in recommending

```python

def get_skills(text):
    nlp = spacy.load("en_core_web_lg")  # Load spaCy English model
    ruler = nlp.add_pipe("entity_ruler")  # Add entity ruler for skill patterns


    ruler.from_disk(skill_pattern_path)
    nlp.pipe_names

    doc = nlp(text)
    myset = []
    subset = []
    for ent in doc.ents:
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    return subset

def unique_skills(x):
    return list(set(x))
!fusermount -u drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

Resume_path = '/content/drive/MyDrive/Resume Recommender System/Resume.csv'
if os.path.exists(Resume_path):
    print("Resume CSV file found!")
else:
    print("Resume CSV file not found. Please check the file path.")



```

Data (Resume) Cleaning and function application on the Resume dataset both cleaning and skills extraction.

```python
clean = []
for i in range(data.shape[0]):
    review = re.sub(
        '(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?"',
        " ",
        data["Resume_str"].iloc[i],
    )
    review = review.lower()
    review = review.split()
    lm = WordNetLemmatizer()
    review = [
        lm.lemmatize(word)
        for word in review
        if not word in set(stopwords.words("english"))
    ]
    review = " ".join(review)
    clean.append(review)

data["Clean_Resume"] = clean

# Extract skills using the get_skills function
data["skills"] = data["Clean_Resume"].str.lower().apply(get_skills)

# Remove duplicates from skill lists
data["skills"] = data["skills"].apply(unique_skills)
data.head()
```

Resume reccomendation logic as well based on user inputs in terms of required skills as well as weights for the respective skills. The initialized weight is 1

```python
required_skills_str = input("Enter required skills (comma-separated): ")
# Initialize with a default weight of 1
required_skills = {skill: 1 for skill in required_skills_str.split(",")}

skill_weights_str = input("Enter weights for each skill (comma-separated): ")
skill_weights = dict(zip(required_skills.keys(), map(float, skill_weights_str.split(","))))

def score_resume(resume_skills, required_skills, skill_weights):
    score = 0
    for req_skill in required_skills:
        weight = skill_weights.get(req_skill, 0) # Get weight, default to 0 if not found
        if req_skill in resume_skills:
            score += weight  # Award weight for exact match

    return score

def rank_resumes(data, required_skills, skill_weights):
    data["score"] = data["skills"].apply(lambda x: score_resume(x, required_skills, skill_weights))
    return data.sort_values(by="score", ascending=False)  # Sort by highest score

ranked_resumes = rank_resumes(data.copy(), required_skills, skill_weights)

print(ranked_resumes[['ID','skills','score']].head(10))  # View top 10 ranked resumes

```
