## Bayesian Linear Regression

```python
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt

# Simulated data
np.random.seed(42)
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 1, 100)

# Bayesian linear regression model
with pm.Model() as model:
    # Priors for unknown model parameters
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta = pm.Normal('beta', mu=0, sigma=10)
    sigma = pm.HalfNormal('sigma', sigma=1)

    # Expected value of outcome
    mu = alpha + beta * x

    # Likelihood
    Y_obs = pm.Normal('Y_obs', mu=mu, sigma=sigma, observed=y)

    # Inference
    trace = pm.sample(2000, return_inferencedata=False)

# Plotting results
pm.plot_trace(trace)
plt.show()
```

## Bayesian Logistic Regression

```python
# Simulated binary data
np.random.seed(42)
x = np.random.normal(0, 1, 100)
y = (np.random.rand(100) < 1 / (1 + np.exp(-(-1 + 2 * x))).astype(int))

# Bayesian logistic regression model
with pm.Model() as model:
    # Priors for unknown model parameters
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta = pm.Normal('beta', mu=0, sigma=10)

    # Logistic regression equation
    mu = pm.math.sigmoid(alpha + beta * x)

    # Likelihood
    Y_obs = pm.Bernoulli('Y_obs', p=mu, observed=y)

    # Inference
    trace = pm.sample(2000, return_inferencedata=False)

# Plotting results
pm.plot_trace(trace)
plt.show()
```

## Hierarchical Modeling

```python
# Simulated hierarchical data
np.random.seed(42)
n_schools = 8
n_students = 10
school_effects = np.random.normal(0, 1, n_schools)
x = np.random.normal(0, 1, (n_schools, n_students))
y = school_effects[:, np.newaxis] + x + np.random.normal(0, 1, (n_schools, n_students))

# Flatten data for modeling
x_flat = x.flatten()
y_flat = y.flatten()
school_indices = np.repeat(np.arange(n_schools), n_students)

# Hierarchical model
with pm.Model() as model:
    # Priors for school effects
    mu_a = pm.Normal('mu_a', mu=0, sigma=10)
    sigma_a = pm.HalfNormal('sigma_a', sigma=1)
    a = pm.Normal('a', mu=mu_a, sigma=sigma_a, shape=n_schools)

    # Model for each student
    mu = a[school_indices] + x_flat
    sigma_y = pm.HalfNormal('sigma_y', sigma=1)

    # Likelihood
    Y_obs = pm.Normal('Y_obs', mu=mu, sigma=sigma_y, observed=y_flat)

    # Inference
    trace = pm.sample(2000, return_inferencedata=False)

# Plotting results
pm.plot_trace(trace)
plt.show()
```
